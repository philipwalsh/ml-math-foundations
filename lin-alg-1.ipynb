```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# a robber gets a 5 minute head start in a car going 150 km/h
# the sheriff drives at 180 km/h
# at what km will the robber be caught
```

![exercise 1](images/alg-1-exercise-01.jpg)

```python
t = np.linspace(0, 40, 1000) # start, finish, n points
distance_robber = 2.5 * t 
distance_sheriff = 3 * (t-5)
```

```python
fig, ax = plt.subplots()
plt.title('Catch the bad guy')
plt.xlabel('time (in min)')
plt.ylabel('distance (in km)')
ax.set_xlim([0, 40])
ax.set_ylim([0, 100])
ax.plot(t, distance_robber, c='red')
ax.plot(t, distance_sheriff, c='green')
plt.axvline(x=30, linestyle='--')
plt.axhline(y=75, linestyle='--')
```
