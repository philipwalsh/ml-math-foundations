```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# a robber gets a 5 minute head start in a car going 150 km/h
# the sheriff drives at 180 km/h
# how long will the robbers poor attempt at a getaway last before he is caught?
# hint: robbers clock starts at zero, shreiff's starts at -5 minutes
```

![exercise 1](images/alg-1-exercise-01.jpg)

```python


t = np.array([0.,40.])         # for this example, its a line, really only need 2 pints
                               #t = np.linspace(0, 40, 1000) something like this if we needed more points
distance_robber  = 2.5 * (t+0) # set the robber at zero
distance_sheriff = 3.0 * (t-5) # lag the sheriff
```

```python
t
```

```python
distance_robber
```

```python
distance_sheriff
```

```python
fig, ax = plt.subplots()
plt.title('Catch the bad hombre')
plt.xlabel('time (in min)')
plt.ylabel('distance (in km)')
ax.set_xlim([0, 40])
ax.set_ylim([0, 100])
ax.plot(t, distance_robber, c='red')
ax.plot(t, distance_sheriff, c='green')
plt.axvline(x=30, linestyle='--')
plt.axhline(y=75, linestyle='--')
```
